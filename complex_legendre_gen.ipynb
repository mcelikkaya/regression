```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import least_squares as mini
import numpy as np
```

```python
#create a sample data for complex datas
#I cloned same method for 2,3 and 4 variable
#I will use this data set for complex fitting
def get_complex_sample3(sample_size=1000):
    x = np.linspace(0, 5, sample_size)
    coeffs = [np.random.rand() for _ in range(3)]
    mult = np.random.randint(1,10, size=4)    
    complex_data = coeffs[0] * np.sin(mult[0] * x) + coeffs[1] * np.sin(mult[1]*x) - coeffs[2] * np.cos(mult[2] * np.cos(mult[3] * x))
    return (x,complex_data)

def get_complex_sample2():
    x = np.linspace(0, 5, 1000)
    coeffs = [np.random.rand() for _ in range(3)]
    mult = np.random.randint(1,10, size=4)    
    complex_data = coeffs[0] * np.sin(mult[0] * x) + coeffs[1] * np.sin(mult[1]*x) 
    return (x,complex_data)

def get_complex_sample4():
    x = np.linspace(0, 5, 1000)
    coeffs = [np.random.rand() for _ in range(4)]
    mult = np.random.randint(1,10, size=4)    
    complex_data = coeffs[3] * np.sin(mult[3]*x) + coeffs[0] * np.sin(mult[0] * x) + coeffs[1] * np.sin(mult[1]*x) - coeffs[2] * np.cos(mult[2] * np.cos(mult[3] * x))
    return (x,complex_data)

    
def draw_degrees(degrees,labels,x,y):
    plt.plot(x, y, c='k', lw=3, label='Actual')
    for index,deg in enumerate(degrees):
        label = labels[index]
        coeffs = np.polynomial.legendre.legfit(x, y, deg)
        plt.plot(x, np.polynomial.legendre.legval(x, coeffs), label=label)
    plt.legend()    
    
    
    
#draw multiple independent graphs in same graph
def plot_multi_complex(title,len):
    for _ in range(len):
        x,y = get_complex_sample3()
        plt.plot(x, y)
    plt.title(title)  
    plt.show()
```

```python
#Combination of multiple data
xs = [ get_complex_sample4() for _ in range(100)]
ys = [ xs[i][1] for i in range(len(xs))]
y_final = np.array(ys).sum(axis=0)
x_final = xs[0][0]

plt.plot(x_final, y_final)
plt.show()
draw_degrees([2,8,30],["Simple","Medium","Complex"],x_final, y_final)
plt.show()
draw_degrees([50,150],["100 var","200 var"],x_final,y_final)
plt.show()
```

```python
#draw multiple independent graphs
plot_multi_complex("Simple Interactions",3)
plot_multi_complex("ComplexInteractions",8)
```
